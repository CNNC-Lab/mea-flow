```python
import pickle
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
from conic_tools.analysis.signals import SpikeList
from conic_tools.analysis.metrics.spikes import compute_isi_stats, compute_spike_stats #, compute_synchrony

from auxiliary import compute_complete_synchrony
```

```python
# Load datasets
data = loadmat('/home/neurobook/Desktop/Research/MEA-data/data/n1-DIV17-01.mat')
ids = data['Channels'][0]
times = data['Times'][0]
spk_times = [(i, times[idx]*1000) for idx, i in enumerate(ids)]
spk_ids = np.unique(ids)
sl_n1 = SpikeList(spk_times, spk_ids)

data = loadmat('/home/neurobook/Desktop/Research/MEA-data/data/n2-DIV17-01.mat')
ids = data['Channels'][0]
times = data['Times'][0]
spk_times = [(i, times[idx]*1000) for idx, i in enumerate(ids)]
spk_ids = np.unique(ids)
sl_n2 = SpikeList(spk_times, spk_ids)

data = loadmat('/home/neurobook/Desktop/Research/MEA-data/data/n3-DIV17-01.mat')
ids = data['Channels'][0]
times = data['Times'][0]
spk_times = [(i, times[idx]*1000) for idx, i in enumerate(ids)]
spk_ids = np.unique(ids)
sl_n3 = SpikeList(spk_times, spk_ids)
```

```python
# Time slice (first 5 seconds, for visualization)
sl_n1_small = sl_n1.time_slice(t_start=0., t_stop=10000.)
sl_n2_small = sl_n2.time_slice(t_start=0., t_stop=10000.)
sl_n3_small = sl_n3.time_slice(t_start=0., t_stop=10000.)
```

```python
results_n1_small = {}
results_n1 = {}
results_n2_small = {}
results_n2 = {}
results_n3_small = {}
results_n3 = {}
```

```python
results_n1_small.update(compute_isi_stats(sl_n1_small, summary_only=False))
results_n1_small.update(compute_complete_synchrony(sl_n1_small, n_pairs=500, time_bin=1., tau=20., time_resolved=False, depth=4))
results_n1_small.update(compute_spike_stats(sl_n1_small, time_bin=50., summary_only=False, display=True))
```

```python
results_n2_small.update(compute_isi_stats(sl_n2_small, summary_only=False))
results_n2_small.update(compute_complete_synchrony(sl_n2_small, n_pairs=500, time_bin=1., tau=20., time_resolved=False, depth=4))
results_n2_small.update(compute_spike_stats(sl_n2_small, time_bin=50., summary_only=False, display=True))
```

```python
results_n3_small.update(compute_isi_stats(sl_n3_small, summary_only=False))
results_n3_small.update(compute_complete_synchrony(sl_n3_small, n_pairs=500, time_bin=1., tau=20., time_resolved=False, depth=4))
results_n3_small.update(compute_spike_stats(sl_n3_small, time_bin=50., summary_only=False, display=True))
```

```python
fig1, ax1 = plt.subplots(1, 4, figsize=(12, 6))
fig1.suptitle("REGULARITY")

ax1[0].set_title(r'$\mathrm{CV_{ISI}}$')
ax1[0].set_xlim([0.7, 1.3])
ax1[0].set_xticklabels(['n1', 'n2', 'n3'])
ax1[0].boxplot([results_n1['cvs'], results_n2['cvs'], results_n3['cvs']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax1[1].set_title(r'$\mathrm{LV_{ISI}}$')
ax1[1].set_xlim([0.7, 1.3])
ax1[1].set_xticklabels(['n1', 'n2', 'n3'])
ax1[1].boxplot([results_n1['lvs'], results_n2['lvs'], results_n3['lvs']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax1[2].set_title(r'$\mathrm{H_{ISI}}$')
ax1[2].set_xlim([0.7, 1.3])
ax1[2].set_xticklabels(['n1', 'n2', 'n3'])
ax1[2].boxplot([results_n1['ents'], results_n2['ents'], results_n3['ents']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax1[3].set_title(r'$\mathrm{ISI_{5\%}}$')
ax1[3].set_xlim([0.7, 1.3])
ax1[3].set_xticklabels(['n1', 'n2', 'n3'])
ax1[3].boxplot([results_n1['isi_5p'], results_n2['isi_5p'], results_n3['isi_5p']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

plt.savefig('./plots/regularity.png')
```

```python
fig2, ax2 = plt.subplots(1, 3, figsize=(10, 6))
fig2.suptitle("ACTIVITY")

ax2[0].set_title(r'$\bar{\nu}$')
ax2[0].set_xlim([0.7, 1.3])
ax2[0].set_xticklabels(['n1', 'n2', 'n3'])
ax2[0].boxplot([results_n1['mean_rates'], results_n2['mean_rates'], results_n3['mean_rates']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax2[1].set_title(r'$\bar{\nu}_{\mathrm{corrected}}$')
ax2[1].set_xlim([0.7, 1.3])
ax2[1].set_xticklabels(['n1', 'n2', 'n3'])
ax2[1].boxplot([results_n1['corrected_rates'], results_n2['corrected_rates'], results_n3['corrected_rates']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

# ax2[2].set_title(r'$\mathrm{Counts}$')
# ax2[2].set_xlim([0.7, 1.3])
# ax2[2].set_xticklabels(['n1', 'n2', 'n3'])
# ax2[2].boxplot([results_n1['counts'], results_n2['counts'], results_n3['counts']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax2[2].set_title(r'$\mathrm{FF}$')
ax2[2].set_xlim([0.7, 1.3])
ax2[2].set_xticklabels(['n1', 'n2', 'n3'])
ax2[2].boxplot([results_n1['ffs'], results_n2['ffs'], results_n3['ffs']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

plt.savefig('./plots/activity.png')
```

```python
fig3, ax3 = plt.subplots(1, 4, figsize=(12, 6))
fig3.suptitle("SYNCHRONY")

ax3[0].set_title(r'$\mathrm{{CC}_{Pearson}}$')
ax3[0].set_xlim([0.7, 1.3])
ax3[0].set_xticklabels(['n1', 'n2', 'n3'])
ax3[0].boxplot([results_n1['ccs_pearson'], results_n2['ccs_pearson'], results_n3['ccs_pearson']], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax3[1].set_title(r'$\mathrm{vanRossum Distance}$')
ax3[1].set_xlim([0.7, 1.3])
ax3[1].set_xticklabels(['n1', 'n2', 'n3'])
ax3[1].boxplot([np.triu(results_n1['d_vr'])[np.nonzero(np.triu(results_n1['d_vr']))].flatten(),
                np.triu(results_n2['d_vr'])[np.nonzero(np.triu(results_n2['d_vr']))].flatten(),
                np.triu(results_n3['d_vr'])[np.nonzero(np.triu(results_n3['d_vr']))].flatten()], 
               notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax3[2].set_title(r'$\mathrm{ISI Distance}$')
ax3[2].set_xlim([0.7, 1.3])
ax3[2].set_xticklabels(['n1', 'n2', 'n3'])
ax3[2].boxplot([results_n1['ISI_distance_matrix'][~np.isnan(results_n1['ISI_distance_matrix'])][1:],
                results_n2['ISI_distance_matrix'][~np.isnan(results_n2['ISI_distance_matrix'])][1:],
                results_n3['ISI_distance_matrix'][~np.isnan(results_n3['ISI_distance_matrix'])][1:]], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

ax3[3].set_title(r'$\mathrm{SPIKE_{Sync} Distance}$')
ax3[3].set_xlim([0.7, 1.3])
ax3[3].set_xticklabels(['n1', 'n2', 'n3'])
ax3[3].boxplot([np.triu(results_n1['SPIKE_sync_matrix'])[np.nonzero(np.triu(results_n1['SPIKE_sync_matrix']))].flatten(), np.triu(results_n2['SPIKE_sync_matrix'])[np.nonzero(np.triu(results_n2['SPIKE_sync_matrix']))].flatten(), np.triu(results_n3['SPIKE_sync_matrix'])[np.nonzero(np.triu(results_n3['SPIKE_sync_matrix']))].flatten()], notch=1, positions=[0.8, 1.0, 1.2], vert=1, sym='')

plt.savefig('./plots/sync.png')
```
